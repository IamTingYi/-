<h3>查詢字串欄位包含某樣式資料</h3>

```python
my_month = "-%02d-" % numpy.random.randint(1, my_last_month + 1) #"-%02d-":不滿兩位數的會自動於前方補0 #%d代表整數
my_month
```

```python
my_monthly = dataframe_2K[dataframe_2K["Date"].str.contains(my_month)] #contains:包含；str.contains(x):字串下包含x的條件式
my_monthly.head(10)
```

```python
my_condition = dataframe_2K["Date"].str.contains(my_month)
my_condition[:10] #看第0~10筆資料 #條件式得到的結果是True or False
```

```python
my_monthly = dataframe_2K[my_condition]
my_monthly.head(10)
```

<h3>計算數值資料欄位平均數及標準差</h3>

```python
my_monly_stats = (my_monthly["Open"].mean(), my_monthly["Open"].std())
my_monly_stats
```

```python
my_monly_stats2 = (my_monthly["Volume"].mean(), my_monthly["Volume"].std())
my_monly_stats2
```

<h2>比較數值資料欄位</h2>

<h3>上界條件查詢</h3>

```python
my_condition = my_monthly["Open"] < my_monly_stats[0] + 0.5 * my_monly_stats[1]
#設定上界增值表(my_condition)為Open的值小於平均+0.5標準差
my_monthly_upper = my_monthly[my_condition]
my_monthly_upper.tail()
```

<h3>下界條件查詢</h3>

```python
my_condition = my_monthly["Open"] > my_monly_stats[0] - 0.5 * my_monly_stats[1]
#設定下界增值表(my_condition)為Open的值大於平均-0.5標準差
my_monthly_lower = my_monthly[my_condition]
my_monthly_lower.head()
```

<h3>複合條件查詢</h3>

```python
my_monthly[
    (my_monthly["Open"] > my_monly_stats[0] - 0.5 * my_monly_stats[1]) &
    (my_monthly["Open"] < my_monly_stats[0] + 0.5 * my_monly_stats[1])
].set_index(["Date"]) #set_index:指派索引為Date
```

```python
my_monthly2 = my_monthly[
    (my_monthly["Volume"] > my_monly_stats2[0] - 0.5 * my_monly_stats2[1]) &
    (my_monthly["Volume"] < my_monly_stats2[0] + 0.5 * my_monly_stats2[1])
]
my_monthly2.set_index(["Date"])
```

<h3>兩資料集交集合</h3>

```python
my_monthly_intersect = my_monthly_upper.merge(my_monthly_lower) #merge():交集 #將upper跟lower取交集
my_monthly_intersect.set_index(["Date"])
```

<h3>兩資料集差集合</h3>

```python
my_monthly[~my_monthly["Date"].isin(my_monthly_intersect["Date"])].set_index(["Date"])
#~:not；isin(x):在x裡面的條件式 #整串條件式的意思為取不在交集裡面的資料
```

<h3>兩資料集聯集合</h3>

```python
my_monthly_union = pandas.concat([my_monthly_intersect, my_monthly2]).drop_duplicates()
#concat:合併資料 #drop_duplicates:把重複的資料刪除 #整串條件式的意思為找符合Open或Volume上下界其中一個條件的資料 
my_monthly_union.set_index(["Date"])
```

<h3>篩選數值資料欄位最大值</h3>

```python
#找出各月份Volumn最大的資料
my_monly_max = None

for mon in range(1, my_last_month + 1):
    _month = "-%02d-" % mon                                        #指派查詢條件為月份
    _df = dataframe_2K[dataframe_2K["Date"].str.contains(_month)]  #各月份的dataframe
    _max = _df[_df["Volume"]==_df["Volume"].max()]                 #抓出各月份Volumn的最大值，不一定只有一筆
    if my_monly_max is None:                                       #如果max裡面沒有東西
        my_monly_max = _max                                        #就把_max丟進去max
    else:                                                          #如果max裡面有東西
        my_monly_max = pandas.concat([my_monly_max, _max])         #就把max跟_max做聯集
        
my_monly_max.sort_values(by="Date")
```

<h3>篩選衍生數值資料欄位最大值</h3>

```python
dataframe_2K["Net"] = numpy.abs(dataframe_2K["Adj Close"] - dataframe_2K["Open"])
#numpy.abs():取絕對值 #建立一個衍伸欄位Net=|Adj Close - Open|

my_monly_max = None

for mon in range(1, 13):
    _month = "-%02d-" % mon
    _df = dataframe_2K[dataframe_2K["Date"].str.contains(_month)]
    _max = _df[_df["Net"]==_df["Net"].max()] 
    if my_monly_max is None:
        my_monly_max = _max
    else:
        my_monly_max = pandas.concat([my_monly_max, _max])
        
my_monly_max.sort_values(by="Date")    
```
