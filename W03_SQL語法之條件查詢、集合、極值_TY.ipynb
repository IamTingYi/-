```python
import pandas
from glob import glob
import numpy
```

<h3>conda install sqldf</h3>

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals()) #快速建立一個公式pysqldf為輸入值q對q做函數sqldf(q, globals())
```

<h3>取得資料檔案群</h3>

```python
data_folder = 'D:/Projects/DB/data'

data_files = glob('%s/IBM-20??.csv' % data_folder)
data_files
```

<h3>隨機載入資料檔</h3>

```python
my_file = numpy.random.randint(len(data_files))
dataframe_2K = pandas.read_csv(data_files[my_file])
```

<h3>顯示部分資料</h3>

```python
dataframe_2K.head()
```

```python
my_last_month_num = dataframe_2K['Date'].iloc[-1]
my_last_month_num = int(my_last_month_num[5:7])
my_last_month_num
```

```python
my_month = '-%02d-' % numpy.random.randint(1, my_last_month_num)
my_monthly = pysqldf("SELECT * FROM dataframe_2K WHERE SUBSTR(Date, 5, 4) = '%s';" % my_month)
#SUBSTR(a,b,c)：抓取a欄位，從第b個位置開始，讀取c格 #抓取月份等於my_month的資料
my_monthly.head(10)
```

<h3>查詢字串欄位包含某樣式資料 (SQL)</h3>

```python
my_Base_SQL = 'SELECT * FROM dataframe_2K'
my_Base_SQL
```

```python
my_month = '-%02d-' % numpy.random.randint(1, 12)
my_monthly = pysqldf("%s WHERE SUBSTR(Date, 5, 4) = '%s';" % (my_Base_SQL, my_month))
#指定第一個%s為my_Base_SQL，第二個%s為my_month
my_monthly.head(10)
```

<h3>計算數值資料欄位平均數及標準差</h3>

```python
my_monly_stats = (my_monthly['Open'].mean(), my_monthly['Open'].std())
my_monly_stats
```

```python
my_monly_stats2 = (my_monthly['Volume'].mean(), my_monthly['Volume'].std())
my_monly_stats2
```

<h3>比較數值資料欄位</h3>

<h3>上界條件查詢</h3>

```python
my_SQL = my_Base_SQL + ' WHERE Open < %s;' #因為Open不是字串，所以%s不用加單引號
my_SQL
```

```python
my_SQL_statement = my_SQL % str(my_monly_stats[0] + 0.5 * my_monly_stats[1]) #str():轉成字串型態
my_SQL_statement
```

```python
my_monthly_upper = pysqldf(my_SQL_statement)
my_monthly_upper.head()
```

<h3>下界條件查詢</h3>

```python
my_SQL = my_Base_SQL + ' WHERE Open > %s;'
my_SQL
```

```python
my_SQL_statement = my_SQL % str(my_monly_stats[0] - 0.5 * my_monly_stats[1])
my_SQL_statement
```

```python
my_monthly_lower = pysqldf(my_SQL_statement)
my_monthly_lower.head()
```

<h3>複合條件查詢</h3>

```python
my_SQL = my_Base_SQL + ' WHERE Open > %s AND Open < %s;'

pysqldf(my_SQL % (
    str(my_monly_stats[0] - 0.5 * my_monly_stats[1]),
    str(my_monly_stats[0] + 0.5 * my_monly_stats[1]))).set_index(['Date']).head()
```

```python
my_SQL = my_Base_SQL + ' WHERE Open < %s OR Open > %s;'

pysqldf(my_SQL % (
    str(my_monly_stats[0] - 0.5 * my_monly_stats[1]),
    str(my_monly_stats[0] + 0.5 * my_monly_stats[1]))).set_index(['Date']).head()
```

```python
my_SQL = my_Base_SQL + ' WHERE Volume > %s AND Volume < %s;'

pysqldf(my_SQL % (
    str(my_monly_stats2[0] - 0.5 * my_monly_stats2[1]),
    str(my_monly_stats2[0] + 0.5 * my_monly_stats2[1]))).set_index(['Date']).head()
```

```python
my_SQL = my_Base_SQL + ' WHERE Volume BETWEEN %s AND %s;' #BETWEEN [下界] AND [上界]，位置不能換

pysqldf(my_SQL % (
    str(my_monly_stats2[0] - 0.5 * my_monly_stats2[1]),
    str(my_monly_stats2[0] + 0.5 * my_monly_stats2[1]))).set_index(['Date']).head()
```

<h3>兩資料集交集合</h3>

```python
my_SQL = my_Base_SQL + ' WHERE (Open BETWEEN %s AND %s) AND (Volume BETWEEN %s AND %s);'

pysqldf(my_SQL % (
    str(my_monly_stats[0] - 0.5 * my_monly_stats[1]),
    str(my_monly_stats[0] + 0.5 * my_monly_stats[1]),
    str(my_monly_stats2[0] - 0.5 * my_monly_stats2[1]),
    str(my_monly_stats2[0] + 0.5 * my_monly_stats2[1]))
).set_index(['Date']).head()
```

<h3>兩資料集差集合</h3>

```python
my_SQL = my_Base_SQL + ' WHERE (NOT Open BETWEEN %s AND %s) OR (NOT Volume BETWEEN %s AND %s);'
#my_SQL = my_Base_SQL + ' WHERE NOT ((Open BETWEEN %s AND %s) AND (Volume BETWEEN %s AND %s));'

pysqldf(my_SQL % (
    str(my_monly_stats[0] - 0.5 * my_monly_stats[1]),
    str(my_monly_stats[0] + 0.5 * my_monly_stats[1]),
    str(my_monly_stats2[0] - 0.5 * my_monly_stats2[1]),
    str(my_monly_stats2[0] + 0.5 * my_monly_stats2[1]))
).set_index(['Date']).head()
```

<h3>兩資料集聯集合</h3>

```python
my_SQL = my_Base_SQL + ' WHERE (Open BETWEEN %s AND %s) OR (Volume BETWEEN %s AND %s);'

pysqldf(my_SQL % (
    str(my_monly_stats[0] - 0.5 * my_monly_stats[1]),
    str(my_monly_stats[0] + 0.5 * my_monly_stats[1]),
    str(my_monly_stats2[0] - 0.5 * my_monly_stats2[1]),
    str(my_monly_stats2[0] + 0.5 * my_monly_stats2[1]))
).set_index(['Date']).head()
```

<h3>篩選數值資料欄位最大值</h3>

```python
my_SQL =  'SELECT SUBSTR(Date, 1, 7) AS YM, MAX(Open) AS Open FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7);'
#GROUP BY：依據指定條件分組
pysqldf(my_SQL).set_index(['YM'])
```

```python
my_SQL = [
    'SELECT * FROM dataframe_2K A',
    'INNER JOIN',
    '(SELECT SUBSTR(Date, 1, 7) AS YM, MAX(Open) AS Open FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7)) B',
    'ON A.Open = B.Open'
]
#依據A的Open欄位inner join B的Open欄位
my_SQL
```

```python
my_SQL = ' '.join(my_SQL)
pysqldf(my_SQL)
```

```python
my_SQL = [
    'SELECT * FROM dataframe_2K A',
    #'SELECT A.Open FROM dataframe_2K A', #取單一欄位的方法
    'INNER JOIN (',
    'SELECT SUBSTR(Date, 1, 7) AS YM, MAX(Volume) AS Volume FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7)) B',
    'ON A.Volume = B.Volume'
]
my_SQL
```

```python
my_SQL = ' '.join(my_SQL)
pysqldf(my_SQL)
```

<h3>篩選衍生數值資料欄位最大值</h3>

```python
dataframe_2K['Net'] = numpy.abs(dataframe_2K['Adj Close'] - dataframe_2K['Open']) #numpy可以解決向量(Vectorization)式的資料型態
dataframe_2K['Net'] = abs(dataframe_2K['Adj Close'] - dataframe_2K['Open'])       #python base只能解決純量式(文字、整數...)的資料型態

my_SQL = [
    'SELECT * FROM dataframe_2K A',
    'INNER JOIN',
    '(SELECT SUBSTR(Date, 1, 7) AS YM, MAX(Net) AS Net FROM dataframe_2K GROUP BY SUBSTR(Date, 1, 7)) B',
    'ON A.Net = B.Net'
]
my_SQL
```

```python
my_SQL = ' '.join(my_SQL)
pysqldf(my_SQL)
```
